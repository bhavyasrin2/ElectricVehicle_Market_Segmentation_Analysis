```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('type_maharashtra.xlsx')
data
```

```python
# Map the "TOTAL" column and "Vehicle Class" column
mapped_data = data[['Unnamed: 23', 'Unnamed: 1']].dropna()

# Print the mapped data
print(mapped_data)
```

```python
data
data.drop(0).dropna()
```

```python
last_row = data.iloc[-1]

# Find the maximum value in the last row
max_value = max(last_row)
```

```python
print("Maximum Value in the Last Row:", max_value)
```

```python
max_value = max(last_row)

# Convert the max_value to an integer
max_value_int = int(max_value)

# Print the integer value of the maximum value
print("Integer Value of Maximum Value:", max_value_int)
```
